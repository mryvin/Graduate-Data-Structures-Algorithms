```python
class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.adjacency_list = [[] for _ in range(vertices)]

    def add_edge(self, u, v):
        self.adjacency_list[u].append(v)
```

```python
from collections import deque

def unique_topological_ordering(self):
        # Initialize in-degrees for all vertices
        in_degree = [0] * self.vertices

        # Calculate in-degrees by traversing the adjacency list
        for vertex in range(self.vertices):
            for neighbor in self.adjacency_list[vertex]:
                in_degree[neighbor] += 1

        # Initialize a queue with vertices having in-degree zero
        queue = deque()
        for vertex in range(self.vertices):
            if in_degree[vertex] == 0:
                queue.append(vertex)

        # Perform topological sorting using Breadth-First Search
        count = 0
        while queue:
            current_vertex = queue.popleft()
            count += 1

            for neighbor in self.adjacency_list[current_vertex]:
                in_degree[neighbor] -= 1
                if in_degree[neighbor] == 0:
                    queue.append(neighbor)

        # Check if the count of visited vertices equals the total number of vertices
        return count == self.vertices

Graph.unique_topological_ordering = unique_topological_ordering
```

```python
# Example WITH unique topological ordering
graph1 = Graph(5)
graph1.add_edge(0, 1)
graph1.add_edge(0, 2)
graph1.add_edge(1, 3)
graph1.add_edge(2, 3)
graph1.add_edge(3, 4)

# Example WITHOUT unique topological ordering
graph2 = Graph(4)
graph2.add_edge(0, 1)
graph2.add_edge(0, 2)
graph2.add_edge(1, 3)
graph2.add_edge(2, 3)
graph2.add_edge(3, 1)

# Example WITH unique topological ordering
graph3 = Graph(6)
graph3.add_edge(0, 1)
graph3.add_edge(0, 2)
graph3.add_edge(1, 3)
graph3.add_edge(2, 3)
graph3.add_edge(3, 4)
graph3.add_edge(4, 5)

# Example WITHOUT unique topological ordering
graph4 = Graph(5)
graph4.add_edge(0, 1)
graph4.add_edge(0, 2)
graph4.add_edge(1, 3)
graph4.add_edge(2, 3)
graph4.add_edge(3, 4)
graph4.add_edge(4, 0)

# Example WITH unique topological ordering
graph5 = Graph(7)
graph5.add_edge(0, 1)
graph5.add_edge(0, 2)
graph5.add_edge(1, 3)
graph5.add_edge(2, 3)
graph5.add_edge(3, 4)
graph5.add_edge(4, 5)
graph5.add_edge(5, 6)

# Example WITHOUT unique topological ordering
graph6 = Graph(4)
graph6.add_edge(0, 1)
graph6.add_edge(1, 2)
graph6.add_edge(2, 3)
graph6.add_edge(3, 1)

#Check for Unique Topological Ordering and Print
result1 = graph1.unique_topological_ordering()
result2 = graph2.unique_topological_ordering()
result3 = graph3.unique_topological_ordering()
result4 = graph4.unique_topological_ordering()
result5 = graph5.unique_topological_ordering()
result6 = graph6.unique_topological_ordering()

print("Graph 1 has a unique topological ordering:", result1)
print("Graph 2 has a unique topological ordering:", result2)
print("Graph 3 has a unique topological ordering:", result3)
print("Graph 4 has a unique topological ordering:", result4)
print("Graph 5 has a unique topological ordering:", result5)
print("Graph 6 has a unique topological ordering:", result6)
```

```python
from collections import deque

graph = {
    1: [2, 3],
    2: [1, 3, 4, 5],
    3: [1, 2, 5, 7, 8],
    4: [2, 5],
    5: [2, 3, 4, 6],
    6: [5],
    7: [3],
    8: [3]
}

visited = set()

def bfs(start):
    queue = deque([start])
    visited.add(start)

    while queue:
        vertex = queue.popleft()
        print(vertex, end=' ')

        for neighbor in graph[vertex]:
            if neighbor not in visited:
                queue.append(neighbor)
                visited.add(neighbor)

# Start BFS from vertex 1
bfs(1)
```
